```python
x = pd.DataFrame()
y = pd.DataFrame()
x['O3'] = comb['O3']
x['DEWP'] = comb['DEWP']
x['month'] = comb['month']
x['day'] = comb['day']
x['RAIN'] = comb['RAIN']
x['WSPM'] = comb['WSPM']
x['wd_num'] = comb.iloc[:, -2]
x['hour'] = comb['hour']
y['TEMP'] = comb['TEMP']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)

print('Mean Squared error :', mean_squared_error(y_test, y_pred))
print('Root Mean Squared error :', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R2 score :', r2_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(x)

# Train the polynomial regression model
model = LinearRegression()
model.fit(X_poly, y)
y_pred = model.predict(poly.transform(x_test))

print('Mean Squared error :', mean_squared_error(y_test, y_pred))
print('Root Mean Squared error :', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R2 score :', r2_score(y_test, y_pred))
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=1.0)  # Regularization strength (alpha can be tuned)
ridge_model.fit(x_train, y_train)

# Make predictions
y_pred = ridge_model.predict(x_test)
print('Mean Squared error :', mean_squared_error(y_test, y_pred))
print('Root Mean Squared error :', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R2 score :', r2_score(y_test, y_pred))
```

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=0.1)  # Regularization strength (alpha can be tuned)
lasso_model.fit(x_train, y_train)

# Make predictions
y_pred = lasso_model.predict(x_test)
print('Mean Squared error :', mean_squared_error(y_test, y_pred))
print('Root Mean Squared error :', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R2 score :', r2_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_model = DecisionTreeRegressor(max_depth = 13)  # Set max_depth to avoid overfitting
tree_model.fit(x_train, y_train)

# Make predictions
y_pred = tree_model.predict(x_test)
print('Mean Squared error :', mean_squared_error(y_test, y_pred))
print('Root Mean Squared error :', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R2 score :', r2_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor()  # Adjust parameters as needed
forest_model.fit(x_train, y_train.values.ravel())  # Use .ravel() to flatten y_train if necessary

# Make predictions
y_pred = forest_model.predict(x_test)
print('Mean Squared error :', mean_squared_error(y_test, y_pred))
print('Root Mean Squared error :', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R2 score :', r2_score(y_test, y_pred))
```

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor(     # Number of trees
    learning_rate=0.1,    # Step size shrinkage
)
xgb_model.fit(x_train, y_train.values.ravel())  # Use .ravel() to flatten y_train if necessary

# Make predictions
y_pred = xgb_model.predict(x_test)

print('Mean Squared error :', mean_squared_error(y_test, y_pred))
print('Root Mean Squared error :', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R2 score :', r2_score(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn_model = KNeighborsRegressor(n_neighbors=5, weights='distance')  # 'uniform' weights or 'distance'
knn_model.fit(x_train, y_train)

# Make predictions
y_pred = knn_model.predict(x_test)

print('Mean Squared error :', mean_squared_error(y_test, y_pred))
print('Root Mean Squared error :', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R2 score :', r2_score(y_test, y_pred))
```
